````
this_data = data_covid[['sex','age','province_of_onset']] # เลือกเฉพาะข้อมูล เพศ,อายุ และจังหวัด จากข้อมูลทั้งหมด และ ใส่ชื่อตารางข้อมูลว่า this_data
````



1.   คำนวณ อายุเฉลี่ย ของผู้ชาย ในข้อมูลทั้งหมด



````
man = this_data[this_data['sex']=='ชาย']  # เลือกเฉพาะผู้ชาย จากตาราง this data และ ใส่ชื่อตารางว่า man
man['age'] # เรียกดูอายุของเพศชาย
````

````
sum = 0 # 
N = 0 # มีคนทั้งหมดกี่คน
for a in man['age']: # ฟังก์ชั่นวนลูปเฉพาะอายุของผู้ชาย
    
    if a > 0: # วนลูปถ้า >0 ก็คือ age
        sum += a # sum = sum + a, sum คือ จำนวนอายุ ถ้า >0 มาบวกกับ a
        N += 1 # จำนวนคน

print(f'อายุเฉลี่ยของผู้ชาย {sum/N}') # print ค่า sum/N ก็จะได้ค่าเฉลี่ยออกมา
````




2.   คำนวณ อายุเฉลี่ย ของผู้ชาย ในจังหวัดขอนแก่น



````
man_KK = man[man['province_of_onset']=='ขอนแก่น'] #เลือกเฉพาะผู้ชายในจังหวัดขอนแก่น จากตาราง this data และใส่ชื่อตารางว่า man_KK

sum = 0
N = 0 # มีคนทั้งหมดกี่คน
for a in man_KK['age']:  # ฟังก์ชั่นวนลูปเฉพาะอายุของผู้ชายในจังหวัดขอนแก่น
    
    if a > 0: # วนลูปถ้า >0 ก็คือ age
        sum += a # sum = sum + a, sum = sum + a  / sum คือ จำนวนอายุ ถ้า >0 มาบวกกับ a
        N += 1 # จำนวนคน

print(f'อายุเฉลี่ยของผู้ชายในจังหวัดขอนแก่น {sum/N}') # print ค่า sum/N ก็จะได้ค่าเฉลี่ยออกมา
````

## Quiz ตัดตารางออกมาเฉพาะปี 2021 ```announce_date``` ในปี 2021 

**Hint**
*   วนลูปหา index ของปี 2021
*   ตัดตารางมาเฉพาะ ปี 2021

````
TF=list()
for each_row in data_covid.iterrows():
    if each_row[1]['announce_date'].split('/')[2] == '2021':
        TF.append(True)
    else:
        TF.append(False)

data_covid[TF].head()
````

### HW กลุ่ม 2 || เลือกข้อมูลมาเฉพาะวัน announce_date เป็นวันที่ 1-20  

````
HW2 = list() #กำหนด ad รับค่า list ว่าง
for i in data_covid.iterrows(): #ใช้คำสั่ง iterrows() วนซ้ำบน data_covid 
  if int(i[1]['announce_date'].split('/')[1]) <= 20: # ใช้ int เพื่อเอาเฉพาะคอลัม announce_date มาคิด และใช้คำสั่ง split เพื่อเอา / ออก และใส่ 1 เพื่อให้เอาแค่เลจวันที่ และใช้ <= 20 เพื่อเอาค่าวันที่ที่ น้อยกว่าและเท่ากับ 20 เท่านั้น
    HW2.append(True) # ถ้าข้อมูลถูกตามที่ใช้คำสั่งให้นับค่า และ แสดงผล
  else:
    HW2.append(False) # ถ้าข้อมูลผิดจะไม่นับค่าและแสดงผล
data_covid[HW2]
````
